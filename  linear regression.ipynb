```python
import pandas as pd
import matplotlib.pyplot as plt

students = ['1', '2', '3', '4', '5']
Computer = [95, 85, 80, 70, 60]
AI = [85, 95, 70, 65, 70]

mean_x = sum( Computer) / len(Computer)
mean_y = sum(AI) / len(AI)
x = [xi - mean_x for xi in Computer]
y = [yi - mean_y for yi in AI]
x2 = [i**2 for i in x]
y2 = [i**2 for i in y]
xy = [x[i] * y[i] for i in range(len(x))]

df = pd.DataFrame({
    'Student': students,
    'C++ (X)': Computer,
    'AI (Y)': AI,
    'x = X - 78': x,
    'y = Y - 77': y,
    'x²': x2,
    'y²': y2,
    'x·y': xy
})

df.loc['Total'] = [''] * 3 + [''] * 2 + [sum(x2), sum(y2), sum(xy)]
print(df)

b = sum(xy) / sum(x2)
a = mean_y - b * mean_x

print(f"\nRegression Equation: Y = {a:.2f} + {b:.3f} * X")
X_new = 45
Y_pred = a + b * X_new
print(f"Predicted AI marks for Computer = 45: {Y_pred:.2f}")

plt.figure(figsize=(8, 6))
plt.scatter(Computer, AI, color='blue', label='Original data')
plt.plot(Computer, [a + b*x for x in Computer], color='red', label='Regression line')
plt.scatter(X_new, Y_pred, color='green', marker='x', s=100, label=f'Prediction (45, {Y_pred:.2f})')
plt.xlabel('Computer Marks')
plt.ylabel('AI Marks')
plt.title('Linear Regression: Predicting AI from Computer')
plt.legend()
plt.grid(True)
plt.show()
```
